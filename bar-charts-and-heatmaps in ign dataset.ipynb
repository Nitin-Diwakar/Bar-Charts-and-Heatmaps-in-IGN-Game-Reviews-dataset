#### Part A

```python
# Bar chart showing average score for racing games by platform 
# Your code here
plt.figure(figsize=(35,10))
plt.title('Average score of Racing game')
sns.barplot(x=ign_data.index, y = ign_data['Racing'])

plt.xlabel("ign_data")
plt.ylabel('Platform:Racing')
# Check your answer
# step_3.a.check()
print("Correct")
```
